```python
import matplotlib.pyplot as plt
```

```python
year_list = ['2016', '2017', '2018', '2019', '2020']
```

```python
x = year_list
```

```python
perf_dict = dict()
```

```python
for year in year_list: 
    perf_dict[year] = dict()
    perf_dict[year]['+14-day Predictions'] = 0.0
    perf_dict[year]['+30-day Predictions'] = 0.0
    perf_dict[year]['Historical Average'] = 0.0
    
```

```python
#setting rmse for three metrics across 5 years, scores generated in model_eval notebook
```

```python
perf_dict['2016']['+14-day Predictions'] = 25.26
perf_dict['2016']['+30-day Predictions'] = 17.40
perf_dict['2016']['Historical Average'] = 40.43
```

```python
perf_dict['2017']['+14-day Predictions'] = 12.35
perf_dict['2017']['+30-day Predictions'] = 17.79
perf_dict['2017']['Historical Average'] = 28.96
```

```python
perf_dict['2018']['+14-day Predictions'] = 11.49
perf_dict['2018']['+30-day Predictions'] = 20.53
perf_dict['2018']['Historical Average'] = 26.05
```

```python
perf_dict['2019']['+14-day Predictions'] = 21.07
perf_dict['2019']['+30-day Predictions'] = 25.77
perf_dict['2019']['Historical Average'] = 64.23
```

```python
perf_dict['2020']['+14-day Predictions'] = 14.47
perf_dict['2020']['+30-day Predictions'] = 19.73
perf_dict['2020']['Historical Average'] = 23.86
```

```python
# perf_dict
```

```python
fig, ax = plt.subplots(figsize=(12,6))

metric_list = ['+14-day Predictions', '+30-day Predictions', 'Historical Average']


for metric in metric_list:
    y_list = []
    for year in perf_dict:
        y_list.append(perf_dict[year][metric])
    ax.plot(year_list, y_list, label=metric)

ax.set_xticks(year_list)
ax.set_ylabel('Prediction Error (RMSE)')
ax.set_title('Comparing RMSE of Historical Average to Model Predictions')
ax.legend()
plt.grid(b=True)
# plt.savefig('graphs/rmse_metrics.jpg', dpi=200)
```
