```python
import matplotlib.pyplot as plt
```

```python
model_list = ['random forest', 'gradient boosting', 'keras sequential', 'simple vgg16', 'custom vgg16']
```

```python
x = list(range(1,len(model_list)+1))
```

```python
x
```

```python
perf_dict = dict()
```

```python
for val in model_list: 
    perf_dict[val] = dict()
    perf_dict[val]['accuracy'] = 0.0
    perf_dict[val]['precision'] = 0.0
    perf_dict[val]['recall'] = 0.0
    perf_dict[val]['f1']= 0.0
    
```

```python
perf_dict['random forest']['accuracy'] = 0.718
perf_dict['random forest']['precision'] = 0.72
perf_dict['random forest']['recall'] = 0.72
perf_dict['random forest']['f1'] = 0.71
```

```python
perf_dict['gradient boosting']['accuracy'] = 0.72
perf_dict['gradient boosting']['precision'] = 0.72
perf_dict['gradient boosting']['recall'] = 0.72
perf_dict['gradient boosting']['f1'] = 0.71
```

```python
perf_dict['keras sequential']['accuracy'] = 0.881
perf_dict['keras sequential']['precision'] = 0.94
perf_dict['keras sequential']['recall'] = 0.85
perf_dict['keras sequential']['f1'] = 0.893
```

```python
perf_dict['simple vgg16']['accuracy'] = 0.905
perf_dict['simple vgg16']['precision'] = 0.94
perf_dict['simple vgg16']['recall'] = 0.89
perf_dict['simple vgg16']['f1'] = 0.914
```

```python
perf_dict['custom vgg16']['accuracy'] = 0.966
perf_dict['custom vgg16']['precision'] = 0.97
perf_dict['custom vgg16']['recall'] = 0.95
perf_dict['custom vgg16']['f1'] = 0.96
```

```python
# perf_dict
```

```python
fig, ax = plt.subplots(figsize=(12,6))

metric_list = ['accuracy', 'precision', 'recall', 'f1']



for metric in metric_list:
    y_list = []
    for val in perf_dict:
        y_list.append(perf_dict[val][metric])
    ax.plot(model_list, y_list, label=metric)

ax.set_xticks(model_list)
ax.legend()
plt.grid(b=True)
plt.savefig('all_models_performance.jpg')
```
